Here 2 means suffering with disease and 1 means not suffering with disease.

Replacing 2 with 1 and 1 with 0, for better understanding.

```python
liver_data['Dataset'] = liver_data['Dataset'].replace([2,1],[1,0])
liver_data['Dataset'].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt  # Import matplotlib

sns.countplot(x='Dataset', data=liver_data)
plt.show()  # This ensures the plot is rendered

```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

sns.countplot(x='Gender', data=liver_data, hue='Dataset')
plt.show()
```

```python
liver_data = pd.get_dummies(liver_data, columns = ['Gender'], drop_first = True)
```

```python
liver_data.head()
```

```python
liver_data['Gender_Male'] = liver_data['Gender_Male'].astype(int)
```

```python
liver_data.head()
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(liver_data.corr(), annot = True)
```

```python
sns.pairplot(liver_data, hue = 'Dataset')
```

```python
liver_data.corr()
```

```python
# X = data[['Albumin_and_Globulin_Ratio', 'Albumin', 'Total_Protiens', 'Aspartate_Aminotransferase', 'Alamine_Aminotransferase', 'Alkaline_Phosphotase', 'Age']]
X = liver_data.drop('Dataset', axis = 1)
y = liver_data['Dataset']
```

```python
X.columns
```

```python
# from imblearn.combine import SMOTETomek
# smk = SMOTETomek(random_state = 42)
# X, y = smk.fit_sample(X,y)
# X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
print("Train Set: ", X_train.shape, y_train.shape)
print("Test Set: ", X_test.shape, y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

# Model Evaluation

# Accuracy Score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, model.predict(X_test))
```

```python
print(f"Accuracy is {round(accuracy_score(y_test, model.predict(X_test))*100,2)}")
```

```python
# Input data
input_data = (72, 3.9, 2.0, 195, 27, 59, 7.3, 2.4, 0.40, 1)

# Convert input data to numpy array and reshape
input_data_as_numpy_array = np.asarray(input_data)
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Feature names (ensure these match the training data)
feature_names = ['Age', 'Total_Bilirubin', 'Direct_Bilirubin', 'Alkaline_Phosphotase',
       'Alamine_Aminotransferase', 'Aspartate_Aminotransferase',
       'Total_Protiens', 'Albumin', 'Albumin_and_Globulin_Ratio',
       'Gender_Male']

# Convert input data to DataFrame
input_data_df = pd.DataFrame(input_data_reshaped, columns=feature_names)

# Make the prediction
prediction = model.predict(input_data_df)
print(prediction)

# Interpret the result
if prediction[0] == 0:
    print('The Person does not have a liver Disease')
else:
    print('The Person has liver Disease')
```
